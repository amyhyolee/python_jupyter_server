#### 04.01. 경기도 공동주택 미분양 현황
- 경기도 시군별 미분양 현황

```python
# 00. 전처리 된 데이터를 load
df = pd.read_csv("./data/"+save_file_name+"_result.csv", encoding='utf8')    # 데이터
# 01. 불필요 컬럼 삭제
# df.drop(columns=['SIGUN_CD'], inplace=True)
# 02.01. 월별 사용량
print("#### 01. 경기도 시군별 미분양 현황")
df = df.groupby(by='SIGUN_NM').sum()
df = df.filter(items=["SIGUN_NM","TOT_LOTOUT_HSHLD_CNT"])
display(df.head(5))

print("\n#### 02. 차트 출력")
df.index.name = '시군'
df.filter(items=["SIGUN_NM","TOT_LOTOUT_HSHLD_CNT"]).plot(kind='line' # line, bar, barh, kde, density
  , title="경기도 시군별 미분양 현황"
  , grid=True
  , legend=True
  , ylim=(0, 3000)
  # , figsize = (20, 20)
)
```
