[문제-03] 저장한 csv 파일을 데이터프렘으로 변환해서 출력하시오.

```python
df = pd.read_csv("data/02_경기데이터드림OpenApi응답명세.csv")
df.head(50)
```

```python
# 문자열 길이 max 구하기
items = ['공공데이터ID', '공공데이터명', '호출URL', '항목ID', '항목명', '항목타입', '요청인자여부']
def get_text_max(name):
    len_max = 0
    df1 = df.filter([name])
    for v in df1.values:
        if len(v[0]) > len_max:
            len_max = len(v[0])
    return len_max
for v in items:
    print(v,' max: ', get_text_max(v))
```

```python
# 1단계: 기본 데이터 삭제
sql = 'truncate table api_tmp'
print(dao.execute(sql))

# 2단계: 데이터 등록
# 등록쿼리
sql = '''
insert into API_TMP (
  INF_ID, INF_NM, CALL_URL, COL_ID, COL_NM, COL_TYPE, REQ_YN
) VALUES (
 %s, %s, %s, %s, %s, %s, %s
)
'''
target_cnt = len(df) # 대상 건수
proc_cnt = 0 # 처리 건수
def save(param): # [동록 / 수정] 단건 처리
    return dao.execute(sql, param)
    
for v in df.values:
    v = tuple(v)
    proc_cnt += save(v)
    
# 3단계: 처리 결과 출력    
print({'대상건수':target_cnt, '처리건수':proc_cnt, '결과': (target_cnt == proc_cnt)})
```

```python
# 1.1. api_mst 데이터 삭제
dao.execute('truncate table api_tmp')
# 1.2. api_mst 등록
dao.execute('''
insert into api_mst(inf_id, inf_nm, call_url) 
  select distinct inf_id, inf_nm, call_url -- distinct: 중복을 제거한다. 
  from api_tmp
;
'''
# 2.1. api_dtl 데이터 삭제
dao.execute('truncate table api_tmp')
# 2.2. api_dtl 등록
dao.execute('''
insert into api_dtl(api_mst_id, COL_ID, COL_NM, COL_TYPE, REQ_YN)
select api_mst_id, COL_ID, COL_NM, COL_TYPE, REQ_YN
  from api_tmp a
 inner join api_mst b on a.inf_id = b.inf_id
;
'''
```
